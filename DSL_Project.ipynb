```python
# df = pd.read_csv('DSL_Winter_Project_2024\development.csv')
df = pd.read_csv('dataset/development.csv')
df
```

```python
df.columns[2:]
```

```python
features = df[df.columns[2:]]
targets = df[['x','y']]
```

```python
pmax = features[[f'pmax[{x}]' for x in range(18)]]
negpmax = features[[f'negpmax[{x}]' for x in range(18)]]
tmax = features[[f'tmax[{x}]' for x in range(18)]]
area = features[[f'area[{x}]' for x in range(18)]]
rms = features[[f'rms[{x}]' for x in range(18)]]
```

```python
pmax.mean(axis=0).plot(kind='bar')
plt.grid()
```

```python
features_list = [pmax,negpmax,tmax,area,rms]

fig, axis = plt.subplots(3,2,figsize=(20,15))

for i,feature in enumerate(features_list):
    feature.mean().plot(kind='bar',ax=axis[i//2,i%2])

plt.title('Means')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
pmax.var(axis=0).plot(kind='bar')
plt.grid()
```

```python
features_list = [pmax,negpmax,tmax,area,rms]

fig, axis = plt.subplots(3,2,figsize=(20,15))

for i,feature in enumerate(features_list):
    feature.var().plot(kind='bar',ax=axis[i//2,i%2])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
pmax = pmax.drop(columns=['pmax[0]','pmax[7]','pmax[12]','pmax[15]','pmax[16]','pmax[17]'])
negpmax = negpmax.drop(columns=['negpmax[0]','negpmax[5]','negpmax[16]','negpmax[17]'])
tmax = tmax.drop(columns=['tmax[0]','tmax[7]','tmax[12]','tmax[15]','tmax[16]','tmax[17]'])
area = area.drop(columns=['area[0]','area[7]','area[12]','area[15]','area[16]','area[17]'])
rms = rms.drop(columns=['rms[16]','rms[17]'])
```

```python
features = pd.concat([pmax,negpmax,tmax,area,rms],axis=1)

features
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features,targets,test_size=0.2,random_state=42)
```

```python

def average_distance(predict : np.ndarray ,actual : np.ndarray ):
  return np.mean(np.sqrt((predict.T[0] - actual.T[0])**2 + (predict.T[1] - actual.T[1])**2))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param_grid = {
    'alpha': [0.1, 0.01, 0.001, 0.0001],
}

rid_model = Ridge(random_state=42)

grid_search = GridSearchCV(rid_model, param_grid=param_grid, n_jobs=-1, verbose=1)

grid_search.fit(X_train, y_train)

print('The best parameters : ',grid_search.best_params_)

rid_model = grid_search.best_estimator_

y_pred = rid_model.predict(X_test)

average_distance(y_test.to_numpy(), y_pred)
```

Since the size of dataset is pretty large, I decide to perfrom the grid search of the RandomForestRegression on only 10000 unique random subset of the dataset.

```python
features
random_samples = np.random.choice(385500, 10000, replace=False)
samples = features.iloc[random_samples]
samples_target = targets.iloc[random_samples]

X_train, X_test, y_train, y_test = train_test_split(samples.to_numpy(), samples_target.to_numpy(), test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features' : ['sqrt','log2'],
    'n_jobs' : [-1]
}

model = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(estimator=model, param_grid=param_grid,verbose=3,cv=None,n_jobs=-1)

grid_search.fit(X_train, y_train)

print('The best parameters for random forest regressor : ',grid_search.best_params_)

best_rf_model = grid_search.best_estimator_

y_pred = best_rf_model.predict(X_test)

average_distance(y_test, y_pred)
```

Now performing the best parameters for whole dataset

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=300,max_features='sqrt',n_jobs=-1,random_state=42)

model.fit(X_train,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
average_distance(y_predictions,y_test.to_numpy())
```
